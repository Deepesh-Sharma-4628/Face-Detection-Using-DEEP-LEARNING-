```python
from mtcnn import MTCNN
```

```python
import cv2
```

```python
detector = MTCNN()
img = cv2.imread('images/fam.jpg')
output = detector.detect_faces(img)
output
for i in output:
    x,y,width,height=i['box']
    cv2.rectangle(img,pt1=(x,y),pt2=(x+width,y+height),color=(0,0,255),thickness=1)


    left_eyeX,left_eyeY=i['keypoints']['left_eye']
    right_eyeX,right_eyeY=i['keypoints']['right_eye']
    noseX,noseY=i['keypoints']['nose']
    mouth_leftX,mouth_leftY=i['keypoints']['mouth_left']
    mouth_rightX,mouth_rightY=i['keypoints']['mouth_right']



    cv2.circle(img,center=(left_eyeX,left_eyeY),color=(0,0,255),thickness=3,radius=2)
    cv2.circle(img,center=(right_eyeX,right_eyeY),color=(0,0,255),thickness=3,radius=2)
    cv2.circle(img,center=(noseX,noseY),color=(0,0,255),thickness=3,radius=2)
    cv2.circle(img,center=(mouth_leftX,mouth_leftY),color=(0,0,255),thickness=3,radius=2)
    cv2.circle(img,center=(mouth_rightX,mouth_rightY),color=(0,0,255),thickness=3,radius=2)


cv2.imshow('window',img)
# cv2.waitKey(0)
```

```python
cap = cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    output = detector.detect_faces(frame)
    for i in output:
        x,y,width,height=i['box']
        cv2.rectangle(frame,pt1=(x,y),pt2=(x+width,y+height),color=(0,0,255),thickness=1)


    cv2.imshow('win',frame)
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break
cv2.destroyAllWindows()
```
